```python
#import library pandas
import pandas as pd
```

```python
#save path dataset ke variabel
dataset = 'melb_data.csv'
#read data dalam bentuk dataframe
dataset = pd.read_csv(dataset)
#menampilkan kolom dataset
dataset.columns
```

```python
#menghapus data yang hilang (na)
dataset = dataset.dropna(axis=0)
```

```python
dataset.describe()
```

```python
#set label
y = dataset.Price
```

```python
#set features
dataset_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = dataset[dataset_features]
```

```python
X.describe()
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
model = DecisionTreeRegressor(random_state=1)

# Fit model
model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(model.predict(X.head()))
```
